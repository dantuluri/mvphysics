```python
import matplotlib.pyplot as plt
import numpy as np

x = [25,50,75,100,125]
y = [1.01,1.4,1.75,2.0,2.27]
plt.plot(x,y, 'ro')
plt.axis([0, 160,0,2.5])

z = np.polyfit(x, y, 2)
f = np.poly1d(z)
print(z)

def func(x):
    return z[0]*x^2+z[1]*x+z[2]

a=[]
b=[]

for x in range(0,160,1):
    y=func(x)
    a.append(x)
    b.append(y)
    

# calculate new x's and y's
x_new = np.linspace(x[0], x[-1])
y_new = f(x_new)

plt.plot(x,y,'ro', x_new, y_new)


plt.show()
```
